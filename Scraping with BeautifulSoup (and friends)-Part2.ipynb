##Part 2: Scraping Le Monde

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import pprint
```

```python
response = requests.get('https://www.lemonde.fr/en')
doc = BeautifulSoup(response.text)
```

```python
rows=[]

items = doc.find_all(class_='article')
for item in items:
    row = {}
    row['headline'] = item.find(class_='article__title').text
    try:
   	    row['subhead'] = item.find(class_='article__desc').text 
    except:
      	row['subhead'] = "no subhead"
    try:
        row['link'] = item.find('a')['href'] 
    except:
        row['link'] = 'no link'
    try:
        row['premium'] = item.find(class_='sr-only').text
    except:
        row['premium'] = 'free'
    try:
        row['author'] = item.find(class_='article__byline').text
    except:
        row['author'] = 'no byline'
    try:
        row['article type'] = item.find(class_='article__type').text
    except:
        row['article type'] = 'normal'
    try:
        row['image link'] = item.find('img')['data-src']
    except:
        row['image link'] = 'no picture'   
    rows.append(row)
rows
```

```python
df = pd.json_normalize(rows)
df
```

```python
df.to_csv('lemonde.csv', index=False)
```
