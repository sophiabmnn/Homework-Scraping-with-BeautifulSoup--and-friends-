#Part 3: Scrape Authorized Third Party Driver License Locations

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations')
doc = BeautifulSoup(response.text)
```

```python
items = doc.select('tr')
items
```

```python
rows = []
list_elements = []
for item in items:
    row = {}
    for elements in item.find_all('td'):
        list_elements.append(elements.text)
    row['company_name'] = list_elements[-4]
    row['address'] = list_elements[-3]
    try:
        row['link'] = item.find('a')['href']
    except:
        row['link'] = 'no link'
    rows.append(row)
rows = rows[1:]
rows
```

```python
df = pd.json_normalize(rows)
df
```

```python
df.to_csv('drivers.csv', index=False) 
```
